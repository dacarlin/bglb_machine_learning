# BglB benchmark 

## Experimental data set

The data set consists of Michaelis-Menten parameters $k_{\mathrm{cat}}$, $K_M$, and $k_{\mathrm{cat}}/K_\mathrm{M}$, and for 100 mutants of $\beta$-glucosidase B from *P. polymxa*.

Experimental data files: 

- `data_set.csv`: the experimental data set 

## Molecular modeling predicts structure 

Modeling consists of modeling each mutation using structural features calculated from molecular modeling software programs Rosetta (45 features) and FoldX (13 features). 

Feature sets: 

- `rosetta_runs/benchmark/feature_set.csv`: feature set used in [first BglB paper](http://dx.doi.org/10.1371/journal.pone.0147596)
- `rosetta_runs/shells/feature_set.csv`: shells protocol 
- `rosetta_runs/ddg_monomer/feature_set.csv`: protocol based on [this protocol capture](https://guybrush.ucsf.edu/benchmarks/captures/DDG)
- `foldx_runs/pssm/feature_set.csv`: 16 features generated by FoldX's PSSM algorithm (runs about 100-fold faster than Rosetta) 

Feature sets in progress: 

- `rosetta_runs/homology_models/feature_set.csv` which consists of homology models that have themselves been run through a EnzRepackMinimize protocol  

## Machine learning maps structure to activity 

Machine learning consists of 10-fold cross-validated predictions from a regularized linear regression model. 

Input files: 

- structural features 
- experimental data 

Outputs: 

- prediction for each of $k_{\mathrm{cat}}$, $K_M$, $k_{\mathrm{cat}}/K_\mathrm{M}$, T$_\mathrm{m}$, and expression for each of 100 mutants 
- associated R values for correlation for each of $k_{\mathrm{cat}}$, $K_M$, $k_{\mathrm{cat}}/K_\mathrm{M}$, T$_\mathrm{m}$, and expression
- plots
