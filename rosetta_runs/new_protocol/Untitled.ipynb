```python
import pandas 
from sqlalchemy import create_engine, inspect 
from glob import glob 

# figure out how the data are structured 
for g in glob( 'out/features*db3' ): 
    db = create_engine( 'sqlite:///{}'.format( g ) )
    inspector = inspect( db ) 
    for table_name in inspector.get_table_names():
        df = pandas.read_sql_table( table_name, db ) 

        print( table_name ) 
        print( df.shape ) 
        print( '-------------------------------------------------------------' )
        print( df.head() )
        print( df.tail() )
        print( '-------------------------------------------------------------' )
        print( df.dtypes ) 
        print( '-------------------------------------------------------------' )
        print( '-------------------------------------------------------------' )
```
